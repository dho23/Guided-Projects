## Introduction

The dataset we will use describes Euro daily exchange rates between 1999 and 2021. The euro (symbolized with €) is the official currency in most of the countries of the European Union.

If the exchange rate of the euro to the US dollar is 1.5, you get 1.5 US dollars if you pay 1.0 euro (one euro has more value than one US dollar at this rate). 

Daria Chemkaeva put together the dataset and made it available on [Kaggle](https://www.kaggle.com/lsind18/euro-exchange-daily-rates-19992020) - the data source is the European Central Bank. Note that the dataset gets regular updates - we downloaded it on February 2022.

Let's start by reading in the dataset and doing some exploration. 

```python
# Reading in dataset

import pandas as pd   # importing pandas library

exchange_rates = pd.read_csv('euro-daily-hist_1999_2020.csv')  # reading file into pandas dataframe
exchange_rates.head()    # first 5 rows to understand structure of dataset
```

```python
exchange_rates.tail()  # last 5 rows
```

```python
exchange_rates.info()
```

```python
exchange_rates.isna().sum().sum()
```

The `exchange_rates` dataset has 5,699 rows and 41 columns. There is 28,767 null values in the dataset. We have 38 variables of the `object` type and 3 variables of the `float` type.  

## Data Cleaning

```python
exchange_rates.rename(columns={'[US dollar ]': 'US_dollar',
                               'Period\\Unit:': 'Time'},
                      inplace=True)                                # renaming
exchange_rates['Time'] = pd.to_datetime(exchange_rates['Time'])    # converting to datetime
exchange_rates.sort_values('Time', inplace=True)                   # Sort by time in ascending
exchange_rates.reset_index(drop=True, inplace=True)                # reset index (dropping initial)
```

In the code above, we did the following:

- Renamed `[US dollar ]` and `Period\Unit:` columns to something easier to type - `US_dollar` and `Time`
- Changed the `Time` column to a `datetime` data type
- Sorted the values by `Time` in ascending order
- Reset the index (and dropped the initial index)

```python
euro_to_dollar = exchange_rates[['Time', 'US_dollar']].copy()      # Isolating Time and US_dollar columns
euro_to_dollar.info()      # Checking 
```

```python
euro_to_dollar['US_dollar'].value_counts()
```

Running the `.value_counts()` method on the `US_dollar` column, we see that we have 62 values of `-`. Let's drop the rows where the `-` character appears in the `US_dollar` column.

```python
euro_to_dollar = euro_to_dollar[euro_to_dollar['US_dollar'] != '-']          # new df excluding '-' characters
euro_to_dollar['US_dollar'] = euro_to_dollar['US_dollar'].astype(float)      # converting column to float
euro_to_dollar.info()
```

## Rolling Mean

Now that we have cleaned the data and removed the `-` characters, we can explore the data further. We'll generate a line plot to visual the evolution of the euro-dollar exchange rate. 

```python
import matplotlib.pyplot as plt
%matplotlib inline 

plt.plot(euro_to_dollar['Time'],
         euro_to_dollar['US_dollar'])
plt.show()
```

Here we see the day-to-day variation in the exchange rate. If we want a smoother visual representation, we can use the rolling mean (also known as the moving average). Using the rolling mean allows us display the long-term trends in the exchange rate. 

For this project, we'll calculate the rolling means for the `US_dollar` column using a moving window of 30 days and then add the rolling means to a new column named `rolling_mean`

```python
euro_to_dollar['rolling_mean'] = euro_to_dollar['US_dollar'].rolling(30).mean()
euro_to_dollar
```

```python
import matplotlib.style as style

style.use('fivethirtyeight')
plt.plot(euro_to_dollar['Time'], euro_to_dollar['rolling_mean'], linewidth=2)
plt.title('30 Day Moving Average')
plt.xlabel('Time')

plt.tick_params(top=False, left=False)
plt.show()
```

## Coming up with an Idea

For this project, we want to show how the euro-dollar rate has changed during the coronavirus pandemic. We can show the 2020 data and the 2016-2019 data as a baseline. We can use subplots to show how the rate changes over the years leading up to the most recent date in the dataset. 

```python
covid_szn = euro_to_dollar.copy(
                   )[(euro_to_dollar['Time'].dt.year >= 2016) & (euro_to_dollar['Time'].dt.year < 2022)]

twenty1619 = covid_szn.copy(
       )[(covid_szn['Time'].dt.year >= 2016) & (covid_szn['Time'].dt.year < 2020)]

twenty20 = covid_szn.copy(
       )[covid_szn['Time'].dt.year >= 2020]
```

```python

### Adding the subplots
plt.figure(figsize=(12, 6))
ax1 = plt.subplot(2,2,1)
ax2 = plt.subplot(2,2,2)
ax3 = plt.subplot(2,1,2)
axes = [ax1, ax2, ax3]



ax1.plot(twenty1619['Time'], twenty1619['rolling_mean'],
         color='#ffa500', linewidth=2.5)
ax1.set_xticklabels([])

ax2.plot(twenty20['Time'], twenty20['rolling_mean'],
         color='#00B2EE', linewidth=2.5)
ax2.set_xticklabels([])

ax3.plot(twenty1619['Time'], twenty1619['rolling_mean'],
         color='#ffa500', linewidth=2.5)
ax3.plot(twenty20['Time'], twenty20['rolling_mean'],
         color='#00B2EE', linewidth=2.5)

ax1.text(735964.0, 1.25, '(2016-2019)', fontsize=18, weight='bold',
        alpha=0.3)
ax2.text(737425.0, 1.23, '(2020-) Covid', fontsize=18, weight='bold',
        alpha=0.3)

### Adding a title and a subtitle
ax1.text(735760.0, 1.325, 'COVID effect on Moving Average of EURO-USD rate',
         fontsize=20, weight='bold')
ax1.text(735760.0, 1.3, '''EURO-USD exchange rates from (2016-2019) and beginning of Covid (2020-)''',
        fontsize=16)

### Adding a signature
ax3.text(735760.0, 0.95, 'DARREN HO' + ' '*110 + 'Source: European Central Bank',
        color = '#f0f0f0', backgroundcolor = '#4d4d4d',
        size=14)

plt.show()
```
